```python
import numpy as np
import numpy.linalg as la
a = np.array([1, 2, 3])
b = np.array([4, 0, 1])
print("Vector a =", a)
print("Vector b =", b)
dot_product = np.dot(a, b)
print("\nDot Product (a . b):", dot_product)
cross_product = np.cross(a, b)
print("Cross Product (a x b):", cross_product)
# L2 Norm (Euclidean Length)
l2_norm = la.norm(a)
print("\nL2 Norm ||a||₂:", l2_norm)

# L1 Norm (Sum of Absolute Values)
l1_norm = la.norm(a, 1)
print("L1 Norm ||a||₁:", l1_norm)

# Infinity Norm (Maximum Absolute Value)
linf_norm = la.norm(a, np.inf)
print("Infinity Norm ||a||∞:", linf_norm)


def cosine(u, v):
    return np.dot(u, v) / (la.norm(u) * la.norm(v))

cos_sim = cosine(a, b)

print("\nCosine Similarity between a and b:",
      round(cos_sim, 3))


print("\n========== SUMMARY ==========")
print("Dot Product       =", dot_product)
print("Cross Product     =", cross_product)
print("L2 Norm           =", round(l2_norm, 3))
print("L1 Norm           =", l1_norm)
print("Infinity Norm     =", linf_norm)
print("Cosine Similarity =", round(cos_sim, 3))
```

```python
import numpy as np
import numpy.linalg as la
# Given matrices
M = np.array([[4., 2.],
              [2., 3.]])
P = np.array([[0., -1.],
              [1.,  0.]])   # 90-degree rotation matrix
inv_M = la.inv(M)
trace_M = np.trace(M)
print("Matrix M:")
print(M)
print("\nInverse of M:")
print(inv_M)
print("\nTrace of M:")
print(trace_M)
# ---------------------------------------------------
# 2. Verify M @ inv(M) = Identity Matrix
# ---------------------------------------------------
identity = np.eye(M.shape[0])  # 2x2 identity matrix
product = M @ inv_M
print("\nM @ inv(M):")
print(product)
print("\nIs M @ inv(M) equal to Identity?")
print(np.allclose(product, identity))
# ---------------------------------------------------
# 3. Matrix Tests
# ---------------------------------------------------
# Symmetric Test: M = Mᵀ
is_symmetric = np.allclose(M, M.T)
print("\nIs M Symmetric?")
print(is_symmetric)
# Orthogonal Test: PᵀP = I
is_orthogonal = np.allclose(P.T @ P, np.eye(P.shape[0]))
print("\nIs P Orthogonal?")
print(is_orthogonal)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Original triangle (3 vertices)
tri = np.array([[0, 1, 0.5],
                [0, 0, 1.0]])

# ---------------------------------------------------
# Function to close the shape for plotting
# ---------------------------------------------------
def close_loop(points):
    return np.hstack([points, points[:, :1]])

# ---------------------------------------------------
# 1. Scaling Matrix (x*2, y*0.5)
# ---------------------------------------------------
S = np.array([[2.0, 0.0],
              [0.0, 0.5]])

# ---------------------------------------------------
# 2. Rotation Matrix (45 degrees)
# ---------------------------------------------------
theta = np.deg2rad(45)

R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

# ---------------------------------------------------
# 3. Apply Scaling then Rotation
# ---------------------------------------------------

# First scale
scaled_tri = S @ tri

# Then rotate
transformed_tri = R @ scaled_tri

# ---------------------------------------------------
# Plot Original and Transformed Triangle
# ---------------------------------------------------

tri_closed = close_loop(tri)
trans_closed = close_loop(transformed_tri)

plt.figure(figsize=(6, 6))

# Original triangle
plt.plot(tri_closed[0], tri_closed[1],
         marker='o', label='Original Triangle')

# Transformed triangle
plt.plot(trans_closed[0], trans_closed[1],
         marker='s', label='Scaled + Rotated')

plt.title("Triangle Transformation")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import numpy.linalg as la

# Given matrix
C = np.array([[4., 1.],
              [2., 3.]])

# ---------------------------------------------------
# 1. Eigenvalues and Eigenvectors
# ---------------------------------------------------

eigenvalues, eigenvectors = la.eig(C)

print("Matrix C:")
print(C)

print("\nEigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)

# ---------------------------------------------------
# 2. Print Eigenvalues
# ---------------------------------------------------

print("\nEigenvalue 1 =", eigenvalues[0])
print("Eigenvalue 2 =", eigenvalues[1])

# ---------------------------------------------------
# 3. Verify C @ v = λv for the first eigenvector
# ---------------------------------------------------

v = eigenvectors[:, 0]      # First eigenvector
lam = eigenvalues[0]        # First eigenvalue

left_side = C @ v
right_side = lam * v

print("\nC @ v =")
print(left_side)

print("\nλ * v =")
print(right_side)

print("\nVerification using np.allclose:")
print(np.allclose(left_side, right_side))
```

```python
import numpy as np
import numpy.linalg as la

# ---------------------------------------------------
# Function to calculate cosine similarity
# ---------------------------------------------------
def cosine(u, v):
    return np.dot(u, v) / (la.norm(u) * la.norm(v))

# ---------------------------------------------------
# Given matrices and vectors
# ---------------------------------------------------
A2 = np.array([[3., 2.],
               [1., 4.]])

b2 = np.array([7., 9.])

cat = np.array([0.9, 0.8, 0.1])
dog = np.array([0.85, 0.7, 0.2])
car = np.array([0.1, 0.2, 0.95])

# ---------------------------------------------------
# 1. Solve A2 x = b2
# ---------------------------------------------------
x = la.solve(A2, b2)

print("Solution x:")
print(x)

# ---------------------------------------------------
# 2. Rank of A2
# ---------------------------------------------------
rank_A2 = la.matrix_rank(A2)

print("\nRank of A2:")
print(rank_A2)

# ---------------------------------------------------
# 3. Cosine Similarities
# ---------------------------------------------------
cos_cat_dog = cosine(cat, dog)
cos_cat_car = cosine(cat, car)

print("\nCosine Similarity (cat, dog):",
      round(cos_cat_dog, 3))

print("Cosine Similarity (cat, car):",
      round(cos_cat_car, 3))

# Determine the most similar pair
if cos_cat_dog > cos_cat_car:
    print("\ncat and dog are more similar.")
else:
    print("\ncat and car are more similar.")
```

```python
import sympy as sp

# Function definition
def g(x):
    return x**3 + 2*x

# 1. Numerical derivative at x = 2 (finite difference)
x0 = 2
h = 1e-6

numerical_derivative = (g(x0 + h) - g(x0)) / h
print("Numerical derivative at x = 2:", numerical_derivative)

# 2. Symbolic derivative of x**3 + 2*x
x = sp.symbols('x')
expr = x**3 + 2*x

symbolic_derivative = sp.diff(expr, x)
print("Symbolic derivative:", symbolic_derivative)

# 3. Evaluate the symbolic derivative at x = 2 and compare
symbolic_value = symbolic_derivative.subs(x, 2)
print("Symbolic derivative at x = 2:", symbolic_value)

# Comparison
print("Difference:", abs(float(symbolic_value) - numerical_derivative))
```

```python
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

h2 = x**2 * y + sp.sin(y)

# 1. dh/dx and dh/dy
dh_dx = sp.diff(h2, x)
dh_dy = sp.diff(h2, y)

print("dh/dx =", dh_dx)
print("dh/dy =", dh_dy)

# 2. Assemble the gradient list
gradient = [dh_dx, dh_dy]
print("Gradient =", gradient)

# 3. Evaluate at (x=2, y=0)
gradient_at = [g.subs({x: 2, y: 0}) for g in gradient]
print("Gradient at (2, 0) =", gradient_at)
```

```python
import sympy as sp

x = sp.symbols('x')

# Expression
y = sp.exp(x**2 + 1)

# Chain rule by hand:
# outer derivative = exp(x**2 + 1)
# inner derivative = 2*x
by_hand = sp.exp(x**2 + 1) * (2*x)

# SymPy differentiation
by_sympy = sp.diff(y, x)

print("By hand   :", by_hand)
print("By SymPy  :", by_sympy)

# Confirm they are identical
print("Match? :", sp.simplify(by_hand - by_sympy) == 0)
```

```python
import numpy as np

# Reproducibility
np.random.seed(42)

# Data
Xb = np.random.randn(6, 4)
Yb = np.random.randn(6, 1)

# Weights
Wa = np.random.randn(4, 8) * 0.1     # input -> hidden
Wb = np.random.randn(8, 1) * 0.1     # hidden -> output

# -------------------------------------------------
# 1. Forward pass: z1, h = ReLU(z1), y_hat, loss
# -------------------------------------------------

z1 = Xb @ Wa                    # Hidden layer pre-activation
h = np.maximum(0, z1)           # ReLU activation

y_hat = h @ Wb                  # Output layer

# Mean Squared Error Loss
loss = np.mean((y_hat - Yb) ** 2)

print("Loss before update =", loss)

# -------------------------------------------------
# 2. Backward pass: dy, dWb, dh, dz1, dWa
# -------------------------------------------------

N = Xb.shape[0]

# dL/dy_hat
dy = (2 / N) * (y_hat - Yb)

# dL/dWb
dWb = h.T @ dy

# dL/dh
dh = dy @ Wb.T

# ReLU derivative
dz1 = dh * (z1 > 0)

# dL/dWa
dWa = Xb.T @ dz1

# -------------------------------------------------
# 3. One gradient-descent step (lr = 0.05)
# -------------------------------------------------

lr = 0.05

Wa_new = Wa - lr * dWa
Wb_new = Wb - lr * dWb

# Forward pass after update
z1_new = Xb @ Wa_new
h_new = np.maximum(0, z1_new)

y_hat_new = h_new @ Wb_new

loss_new = np.mean((y_hat_new - Yb) ** 2)

print("Loss after update  =", loss_new)
print("Loss decreased?    =", loss_new < loss)

# Optional: Print gradient shapes
print("\nGradient Shapes:")
print("dy  :", dy.shape)
print("dWb :", dWb.shape)
print("dh  :", dh.shape)
print("dz1 :", dz1.shape)
print("dWa :", dWa.shape)
```

```python
import sympy as sp

# -------------------------------------------------
# 1. Hessian of x**4 + y**2
# -------------------------------------------------

x, y = sp.symbols('x y')

f = x**4 + y**2

H = sp.hessian(f, (x, y))

print("Hessian:")
sp.pprint(H)

# -------------------------------------------------
# 2. Gradient descent to minimise (x - 7)**2
# -------------------------------------------------

xv = 0.0
lr = 0.1

for step in range(20):
    grad = 2 * (xv - 7)     # f'(x)
    xv = xv - lr * grad

print("\nFinal x:", xv)
```
